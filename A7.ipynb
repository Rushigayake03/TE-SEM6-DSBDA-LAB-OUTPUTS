# Tokenization

Tokenization is the process of splitting a sentence into individual words called tokens.

Example:
"Machine Learning is fun" → ["Machine", "Learning", "is", "fun"]

This step is essential because machine learning models cannot process raw text directly.

sentences = sent_tokenize(text)

# POS Tagging

Part-of-Speech (POS) tagging assigns grammatical categories to each word.

Examples:
- NN → Noun
- VB → Verb
- JJ → Adjective

This helps in understanding the role of each word in a sentence.

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

# Stopwords Removal

Stopwords are commonly used words such as "is", "the", "and" which do not carry significant meaning.

Removing stopwords helps:
- Reduce data size
- Improve model performance
- Focus on important words

We also ensure:
- Words are converted to lowercase
- Only alphabetic words are kept

```python
text="How to remove stop words with NLTK library in Python7"
text= re.sub('[^a-zA-Z]', ' ', text)
tokens= word_tokenize(text.lower())
filtered_text= []
for w in tokens:
    if w not in stop_words:
        filtered_text.append(w)
print("Tokenized Sentence:", tokens)
print("Filtered Sentence:", filtered_text)
```

# Stemming

Stemming reduces words to their root form by removing suffixes.

Examples:
- playing → play
- running → run

However, stemming may produce non-dictionary words.

Example:
- "studies" → "studi"

```python
ps = PorterStemmer()
stemmed_words = ['wait', 'waiting', 'waited', 'waits']
for w in stemmed_words:
    print(w, ":", ps.stem(w))
print("Original Sentence:", stemmed_words)
```

# Lemmatization

Lemmatization converts words into their meaningful base form using vocabulary and grammar.

Examples:
- better → good
- running → run

It is more accurate than stemming because it considers context.

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in filtered_text]

print("Lemmatized Words:", lemmatized_words)
```

# TF-IDF (Term Frequency - Inverse Document Frequency)

TF-IDF is used to convert text into numerical form.

TF (Term Frequency):
Measures how frequently a word appears in a document.

IDF (Inverse Document Frequency):
Measures how important a word is across multiple documents.

TF-IDF = TF × IDF

Words with high TF-IDF scores are considered important.

This representation is widely used in text mining and information retrieval.

```python
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(filtered_text).toarray()
print(X)
```
