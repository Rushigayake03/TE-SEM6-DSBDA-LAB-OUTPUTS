## df1['Species'].unique()

Purpose:
Returns all unique (distinct) values present in the "Species" column.

Explanation:
- The "Species" column contains categorical data.
- The `unique()` function scans the column and extracts all distinct values.
- The output is a NumPy array of unique species names.

Use Case:
Used to identify how many categories exist in the dataset and what they are.

---

## df1.groupby(['Species']).median()

Purpose:
Calculates the median (middle value) of each numerical column for every species.

Explanation:
- `groupby(['Species'])` splits the dataset into groups based on species.
- `median()` computes the middle value for each numerical feature within each group.
- The result is a table where each row corresponds to a species and each column shows the median value.

Use Case:
Useful when data contains outliers, as median is more robust than mean.

---

## df1.groupby(['Species']).std()

Purpose:
Calculates the standard deviation of each numerical column for every species.

Explanation:
- `groupby(['Species'])` divides the dataset into species-wise groups.
- `std()` measures how much the values in each group vary from the mean.
- Higher standard deviation indicates more spread in the data.

Use Case:
Helps understand variability of features within each species.
