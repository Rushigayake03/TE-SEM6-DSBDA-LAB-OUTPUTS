# Import Libraries

In this step, we import the necessary libraries for data analysis and visualization.

- pandas:
  Used for handling and analyzing tabular data.

- seaborn:
  A statistical data visualization library built on top of matplotlib. It provides attractive and informative graphs.

- matplotlib.pyplot:
  Used for plotting graphs and customizing visualizations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

The Titanic dataset is loaded using seaborn's built-in function load_dataset().

This dataset contains information about passengers such as:
- Age
- Gender
- Class
- Fare
- Survival status

It is commonly used for data analysis and machine learning tasks.

```python
df = sns.load_dataset('titanic')

print(df.head())
```

# Check Missing Values

Before performing analysis, it is important to check for missing values.

Missing data can affect:
- Accuracy of analysis
- Visualization results

The isnull().sum() function is used to count missing values in each column.

```python
print("\nMissing Values:\n", df.isnull().sum())
```

# Pattern Analysis Using Seaborn

Seaborn is used to visualize patterns in the dataset.

Count plots are used to show the distribution of categorical variables.

1. Survival Count:
   Shows how many passengers survived vs not survived.

2. Survival by Gender:
   Helps analyze whether gender influenced survival.

3. Survival by Class:
   Helps understand the impact of passenger class on survival.

These visualizations help in identifying important trends in the dataset.

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='class', hue='survived', data=df)
plt.title("Survival by Class")
plt.show()
```

# Fare Distribution (Histogram)

A histogram is used to show the distribution of a numerical variable.

In this case, we plot the distribution of ticket fare.

- sns.histplot():
  Used to create histogram using seaborn.

- bins:
  Defines the number of intervals.

- kde=True:
  Adds a smooth curve to visualize distribution.

This helps in understanding:
- Range of fare values
- Most common fare ranges
- Outliers in the data

```python
sns.histplot(df['fare'], bins=30, kde=True)
```
