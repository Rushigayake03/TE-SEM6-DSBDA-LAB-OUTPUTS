# Approach

This assignment implements Naïve Bayes classification on the Iris dataset.

Steps:
1. Load dataset
2. Check missing values
3. Encode categorical data
4. Split dataset
5. Train Gaussian Naïve Bayes model
6. Make predictions
7. Evaluate using confusion matrix and metrics

# Import Required Libraries

In this step, we import all necessary Python libraries required for data handling, model building, and evaluation.

- pandas:
  Used for loading and manipulating structured datasets. It provides DataFrame objects which are easy to work with.

- train_test_split (from sklearn.model_selection):
  Used to divide the dataset into training and testing sets. This ensures that the model is evaluated on unseen data.

- LabelEncoder (from sklearn.preprocessing):
  Machine learning algorithms require numerical input. Since the 'Species' column contains categorical values (like Setosa, Versicolor), LabelEncoder converts them into numeric labels.

- GaussianNB (from sklearn.naive_bayes):
  This is the implementation of the Naïve Bayes algorithm for continuous data. It assumes that features follow a normal (Gaussian) distribution.

- confusion_matrix:
  Used to evaluate classification results by comparing actual and predicted values.

- classification_report:
  Provides detailed metrics such as precision, recall, F1-score for each class.

- accuracy_score:
  Calculates overall accuracy of the model.

```python

import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

# Load Dataset

Load Iris dataset using pandas.

```python

df = pd.read_csv('Datasets/Iris.csv')
df
```

# Check Missing Values

Check if dataset contains null values.


df.isnull().sum()

# Label Encoding

Machine learning models cannot work directly with categorical (text) data. Therefore, we need to convert the 'Species' column into numerical values.

Label Encoding assigns a unique integer to each category:

- Setosa → 0
- Versicolor → 1
- Virginica → 2

The LabelEncoder class from sklearn.preprocessing is used for this purpose.

Steps:
1. Create an instance of LabelEncoder.
2. Apply fit_transform() on the 'Species' column.
   - fit() learns the mapping
   - transform() applies the mapping

After encoding, the dataset becomes fully numeric and suitable for model training.

```python

label_encoder = LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])
df
```

# Define Features (X) and Target (y)

In supervised machine learning, the dataset is divided into:

- Features (X):
  These are the input variables used by the model to make predictions.
  In this case:
  - SepalLength
  - SepalWidth
  - PetalLength
  - PetalWidth

- Target (y):
  This is the output variable that we want to predict.
  In this case:
  - Species

We separate the dataset accordingly using:
- drop() to remove target column from features
- Direct selection for target column

```python
x = df.drop('Species', axis=1)
y = df['Species']
```

# Train-Test Split

To evaluate the performance of the model, the dataset is split into two parts:

- Training Set:
  Used to train the model (learn patterns)

- Testing Set:
  Used to evaluate the model on unseen data

The function train_test_split() is used for this purpose.

Parameters:
- test_size = 0.2 → 20% data is used for testing
- random_state = 0 → ensures reproducibility (same split every time)

This step is crucial to avoid overfitting and to ensure that the model generalizes well.

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0)
```

# Train Naïve Bayes Model

Naïve Bayes is based on Bayes' Theorem:

P(A|B) = (P(B|A) * P(A)) / P(B)

It calculates the probability of a class given the input features.

We use Gaussian Naïve Bayes (GaussianNB) because:
- The features in Iris dataset are continuous
- It assumes data follows a normal distribution

Steps:
1. Create model instance
2. Train model using fit()

The model learns probability distributions of features for each class.

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

# Prediction

After training the model, we use it to make predictions on the test dataset.

The predict() function:
- Takes input features (x_test)
- Returns predicted class labels

These predictions are then compared with actual values to evaluate model performance.

```python
y_pred = model.predict(x_test)
```

```python
model.predict([[1,5.1, 3.5, 1.4, 0.2]])
```

```python
model.predict([[2, -0.8, 1.2, -1.5, 0.3]])
```

# Confusion Matrix

A confusion matrix is used to evaluate classification performance.

For multi-class classification, it is a square matrix where:

- Rows represent actual classes
- Columns represent predicted classes

Each element represents:
Number of instances where actual class = i and predicted class = j

It helps in identifying:
- Correct predictions
- Misclassifications

From this matrix, we can derive:
- True Positive (TP)
- False Positive (FP)
- True Negative (TN)
- False Negative (FN)

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# Classification Report

The classification report provides detailed evaluation metrics:

- Precision:
  Out of all predicted positives, how many are correct.

- Recall:
  Out of all actual positives, how many were correctly predicted.

- F1-score:
  Harmonic mean of precision and recall.

- Support:
  Number of actual samples for each class.

This report gives a deeper understanding of model performance beyond accuracy.

```python
print(classification_report(y_test, y_pred))
```

# Accuracy

Accuracy is the ratio of correctly predicted instances to total instances.

Formula:
Accuracy = (Number of correct predictions) / (Total predictions)

It is calculated using accuracy_score() function.

Although accuracy is useful, it should be used along with other metrics for better evaluation.

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
