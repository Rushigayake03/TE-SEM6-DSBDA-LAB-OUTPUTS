# Data Wrangling II - Academic Performance Dataset

This experiment focuses on handling missing values, detecting and treating outliers, and applying data transformations to improve data quality and distribution.

```python
import pandas as pd
import numpy as np

```

```python
data = {
    "Name":["abc","lmn","stu","xyz","efg","klm","rst","uvw",np.nan,"def"],
    "Maths":[96,88,np.nan,73,44,88,72,84,93,92],
    "Reading":[77,200,88,np.nan,np.nan,91,78,np.nan,84,93],
    "Writing":[86,np.nan,170,87,69,91,np.nan,88,92,83],
    "Placement":[88,76,91,84,96,84,80,78,np.nan,93],
    "Gender":["Male","Female",np.nan,"Female","Female","Male","Female", "Male", "Male", "Male"],
    "Joining Date":[2018,2019,2022,2018,2017,2023,2016,2020,2020,2019],
    "Region":["Pune","Mumbai","Delhi",np.nan,"Surat","Solapur","Bengaluru","Kolkata",np.nan,"Pune"]
}

df = pd.DataFrame(data)
df
```

## 1. Checking Missing Values and Inconsistencies

Missing values are detected using `isnull()` function.

```python
df.isnull().sum()
```

## Handling Missing Values in Numerical Columns

In this step, missing values in numerical columns are handled using **mean imputation**.

- The columns **Maths, Reading, Writing, and Placement** are selected as they contain numerical data.
- The `fillna()` function is used to replace missing values.
- The missing values are replaced with the **mean (average)** of their respective columns.

### Why Mean Imputation?
- It maintains the overall distribution of the data  
- Simple and effective for numerical datasets  
- Prevents data loss compared to dropping rows  

Thus, all missing numerical values are replaced with meaningful values.

```python
num_cols = ["Maths","Reading","Writing","Placement"]

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
df
```

## Handling Missing Values in Categorical Columns

In this step, missing values in categorical columns are handled using **mode imputation**.

- The columns **Gender** and **Region** are selected as they contain categorical (non-numeric) data.
- A loop is used to process each column individually.
- The `fillna()` function replaces missing values with the **mode** (most frequent value) of that column.

### Why Mode Imputation?
- Suitable for categorical data where mean/median cannot be applied  
- Replaces missing values with the most common category  
- Helps maintain consistency in the dataset  

Thus, missing categorical values are filled with the most frequent category in each column.

```python
cat_cols = ["Gender","Region"]

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

df
```

## 2. Outlier Detection using IQR Method

In this step, outliers are detected using the **Interquartile Range (IQR) method**.

### Steps Performed:

- **Q1 (First Quartile)**: Represents the 25th percentile of the data  
- **Q3 (Third Quartile)**: Represents the 75th percentile of the data  
- **IQR (Interquartile Range)**: Calculated as  
  IQR = Q3 − Q1  
  It represents the range of the middle 50% of the data  

### Outlier Detection Rule:

- Lower Bound = Q1 − 1.5 × IQR  
- Upper Bound = Q3 + 1.5 × IQR  

Any value:
- Less than the lower bound, or  
- Greater than the upper bound  

is considered an **outlier**.

### Output:

- The result is a boolean DataFrame:
  - **True** → indicates an outlier  
  - **False** → indicates a normal value  

This method helps in identifying extreme values that may affect data analysis.

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR)))

outliers
```

```python
df
```

## Outlier Treatment using Capping (Clipping)

After detecting outliers, they are handled using the **capping method** with the help of the `clip()` function.

### Steps Performed:

- The **lower bound** is calculated as:  
  Lower Bound = Q1 − 1.5 × IQR  

- The **upper bound** is calculated as:  
  Upper Bound = Q3 + 1.5 × IQR  

### Outlier Treatment:

- Values below the lower bound are replaced with the lower bound  
- Values above the upper bound are replaced with the upper bound  
- Values within the range remain unchanged  

### Why Clipping?

- Prevents extreme values from affecting analysis  
- Retains all data points (no data loss)  
- Maintains consistency in the dataset  

The `clip()` function applies these bounds column-wise (using `axis=1`), ensuring each numerical column is treated independently.

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df[num_cols] = df[num_cols].clip(lower=lower, upper=upper, axis=1)

df
```

## 3. Data Transformation

Data transformation is applied to improve the distribution of data.  
Here, log transformation is used to reduce skewness and make the data more normally distributed.

```python
df["Maths_log"] = np.log(df["Maths"])
df
```

### Reason for Transformation

- Log transformation reduces skewness in the data  
- Helps in converting non-linear relationships into linear ones  
- Improves performance of machine learning models
