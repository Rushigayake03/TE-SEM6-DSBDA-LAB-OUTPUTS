# Import Libraries

In this step, we import the required libraries for data visualization.

- pandas:
  Used for handling dataset.

- seaborn:
  Used for creating statistical visualizations like box plots.

- matplotlib.pyplot:
  Used to display and customize plots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

The Titanic dataset is loaded using seaborn's built-in function.

This dataset contains information about passengers such as:
- Age
- Gender (sex)
- Survival status
- Class
- Fare

It is widely used for data analysis and visualization.

```python
df = sns.load_dataset('titanic')

print(df.head())
```

# Handling Missing Values

The 'age' column contains missing values.

Missing values can affect visualization and analysis.

We check missing values using:
df['age'].isnull().sum()

Seaborn automatically ignores missing values in plots.

```python
print(df['age'].isnull().sum())
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df
```

# Box Plot

A box plot is used to visualize the distribution of a numerical variable.

In this assignment:
- x-axis → Gender (sex)
- y-axis → Age
- hue → Survival status

The box plot shows:
- Median (middle line)
- Quartiles (box)
- Outliers (points outside whiskers)

This helps in comparing age distribution across gender and survival categories.

```python
sns.boxplot(x = 'sex', y = 'age', data = df)
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

# Explanation of Box Plot: Age Distribution by Gender and Survival

The given box plot represents the distribution of passenger ages based on gender (male, female) and survival status (0 = not survived, 1 = survived).

---

##  Understanding the Box Plot Components

Each box plot consists of:

- Median (center line): Represents the middle value of age.
- Interquartile Range (IQR): The box shows the range between 25th percentile (Q1) and 75th percentile (Q3).
- Whiskers: Represent the range of data excluding outliers.
- Outliers: Data points that lie far from the normal distribution.

---

##  Male Passengers

###  Not Survived (Blue)
- Median age is around 28–30 years.
- Age distribution is wider, indicating more variation.
- Presence of several outliers, especially older passengers (above 70).

###  Survived (Orange)
- Median age is slightly lower than non-survivors.
- Age range is slightly narrower.
- Fewer older passengers survived compared to non-survivors.

Interpretation:
Male passengers had a lower survival rate overall, and survival does not strongly favor older males.

---

##  Female Passengers

###  Not Survived (Blue)
- Median age is around 24–26 years.
- Distribution is moderately spread.
- Fewer outliers compared to males.

###  Survived (Orange)
- Median age is slightly higher than non-survivors (~27–30 years).
- Wider spread indicates survival across different age groups.

 Interpretation:
Female passengers had a higher survival rate, and survival was observed across a broader age range.

---

##  Key Observations

1. Females had a significantly higher survival rate compared to males.
2. Male non-survivors show more extreme age values (more outliers).
3. Female survivors are distributed across a wider age range, indicating better survival chances.
4. Younger passengers appear slightly more likely to survive, especially among females.
5. The variability in age is higher among males than females.

---

##  Conclusion

The box plot clearly shows that gender played a crucial role in survival during the Titanic disaster. Females had better survival chances across different age groups, while male survival was limited and more concentrated in specific age ranges.

This supports the historical understanding of the "women and children first" evacuation policy.
