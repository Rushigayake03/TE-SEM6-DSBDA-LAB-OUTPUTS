# Import Libraries

We import the required libraries for data manipulation and machine learning:

- pandas → for handling dataset
- numpy → for numerical operations
- matplotlib → for visualization (optional)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Dataset

The dataset is loaded using pandas read_csv() function.

- The dataset file should be present inside the 'datasets' folder.
- head() is used to display the first 5 rows.

```python
df = pd.read_csv("Datasets/Social_Network_Ads.csv")
df.head()
```

# Data Preprocessing

Steps performed:

1. Remove 'User ID' column as it is not useful for prediction.
2. Convert categorical column 'Gender' into numeric using one-hot encoding.

pd.get_dummies() is used to convert categorical values into binary columns.

# Define Features and Target

- X → Independent variables (Age, Salary, Gender)
- y → Dependent variable (Purchased)

The model learns from X to predict y.

```python
x = df.drop(['User ID', 'Purchased', 'Gender'], axis=1)
x
```

```python

y = df['Purchased']
y
```

# Train-Test Split

The dataset is divided into:

- Training data (80%) → used to train model
- Testing data (20%) → used to evaluate model

random_state ensures reproducibility.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

# Feature Scaling

Feature scaling standardizes the data so that all features contribute equally.

StandardScaler transforms data such that:
- Mean = 0
- Standard deviation = 1

This improves performance of Logistic Regression.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Train Logistic Regression Model

Logistic Regression is used for binary classification.

- fit() trains the model on training data
- It learns relationship between features and target

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python

y_test = classifier.predict(X_test)
```

```python

y_train
```

```python
y_test
```

# Make Predictions

The trained model is used to predict values on test data using predict() function.

```python
y_pred = classifier.predict(X_test)
y_pred
```

# Confusion Matrix

Confusion matrix shows prediction results:

              Predicted
              1     0
                
Actual 1    -------        TP --------  FN

Actual 0     --------       FP --------    TN

Where:
- TP → Correct positive prediction
- TN → Correct negative prediction
- FP → Incorrect positive prediction
- FN → Incorrect negative prediction

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python

score = accuracy_score(y_test, y_pred)
score
```

```python
print(classification_report(y_test, y_pred))
```

```python
print('True Positive:', cm[0][0])
print('True Negative:', cm[1][1])
print('False Positive:', cm[0][1])
print('False Negative:', cm[1][0])
print('Accuracy:', score)
print('Error Rate:', 1-score)
precision = cm[0][0]/(cm[0][0]+cm[0][1])
print('Precision:', precision)
```

# Evaluation Metrics

Formulas:

- Accuracy = (TP + TN) / TP + TN + FP + FN
- Error Rate = 1 - Accuracy
- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)

These metrics evaluate model performance.
