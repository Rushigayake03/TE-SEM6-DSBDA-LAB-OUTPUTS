## 1. Importing Libraries

The code begins by importing the necessary libraries required for data handling and visualization.

- pandas:
  Used for data manipulation and handling tabular datasets.

- seaborn:
  A visualization library built on top of matplotlib, used for creating statistical graphics.

- matplotlib.pyplot:
  Used for creating plots and controlling figure layout.

These libraries together allow efficient data analysis and visualization.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Loading the Dataset

The dataset is loaded using:

    iris = sns.load_dataset('iris')

This loads the Iris dataset into a DataFrame named `iris`.

The dataset contains the following columns:
- sepal_length
- sepal_width
- petal_length
- petal_width
- species

The first four columns are numerical features, and the last column is categorical.


```python
df = sns.load_dataset('iris')

print(df.head())
```

## 4. Identifying Features and Their Types

is used to list all features along with their data types.

Observations:
- sepal_length → float (numeric)
- sepal_width → float (numeric)
- petal_length → float (numeric)
- petal_width → float (numeric)
- species → object (categorical)

This step is important to determine which type of visualization should be applied to each feature.


```python
print("\nFeature Types:\n")
df.dtypes
```

## 5. Histogram Visualization

Explanation:

- plt.subplots(2, 2):
  Creates a grid of 4 plots (2 rows × 2 columns) to display all histograms together.

- sns.histplot():
  Plots the histogram of each feature.

- kde=True:
  Adds a smooth curve (Kernel Density Estimate) to visualize the distribution shape.

- ax parameter:
  Assigns each plot to a specific position in the grid.

Purpose:
Histograms are used to analyze:
- Distribution of data
- Skewness (left or right)
- Spread of values
- Presence of multiple clusters


```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sns.histplot(df['sepal_length'], kde=True, ax=axes[0, 0])
sns.histplot(df['sepal_width'], kde=True, ax=axes[0, 1])
sns.histplot(df['petal_length'], kde=True, ax=axes[1, 0])
sns.histplot(df['petal_width'], kde=True, ax=axes[1, 1])

plt.tight_layout()
plt.show()
```

## 6. Boxplot Visualization


- Each boxplot represents distribution of a feature grouped by species.

- x='species':
  Groups the data based on the type of flower.

- y='feature':
  Shows the numerical values of the feature.

- Boxplot Components:
  - Median (middle line)
  - Quartiles (box edges)
  - Whiskers (range of data)
  - Outliers (points outside whiskers)

Purpose:
Boxplots are used to:
- Compare distributions across categories
- Identify spread and variability
- Detect outliers
- Observe differences between species


```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
sns.boxplot(y='petal_length', x = 'species', data = df, ax=axes[0, 0])
sns.boxplot(y='petal_width', x = 'species', data = df, ax=axes[0, 1])
sns.boxplot(y='sepal_length', x = 'species', data = df, ax=axes[1, 0])
sns.boxplot(y='sepal_width', x = 'species', data = df, ax=axes[1, 1])
```

# Comparison of Distributions and Identification of Outliers

The distributions of the features were analyzed using histograms and boxplots.

---

## 🔹 Comparison of Feature Distributions

### 1. Sepal Length
- The histogram shows a roughly **normal distribution** with slight right skew.

Interpretation:
Sepal length does not clearly separate species due to overlapping distributions.

---

### 2. Sepal Width

### 3. Petal Length

### 4. Petal Width

Petal width is one of the most important features for classification.

## 🔹 Identification of Outliers

Outliers are values that lie far from the general distribution and are visible as points outside the whiskers in boxplots.

- **Sepal Width**:
  - Contains noticeable outliers on both lower and upper ends.
  - Indicates high variability in this feature.

- **Sepal Length**:
  - A few mild outliers are present.
  - Most values are within a consistent range.

- **Petal Length**:
  - Very few outliers.
  - Data is well-clustered.

- **Petal Width**:
  - Minimal or no significant outliers.
  - Distribution is tightly grouped.

---

## 🔹 Final Conclusion

- **Petal Length and Petal Width** show clear separation and are the most useful features for distinguishing species.
- **Sepal Length and Sepal Width** show overlapping distributions and higher variability.
