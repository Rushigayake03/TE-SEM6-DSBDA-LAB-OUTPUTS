## 6. Checking Missing Values

The `isnull().sum()` function is used to identify missing values in each column.

df.isnull().sum()

```python
df.isnull()
```

## 7. Data Type Conversion

The 'Gender' column is converted from object (string) type to category type.

- This improves memory efficiency
- Helps in better handling of categorical data

```python
df['Gender'] = df['Gender'].astype('category')
```

## 8. Checking Data Types

The `dtypes` function displays the data type of each column in the dataset.

```python
df.dtypes
```

## 9. Data Normalization

Normalization is performed to scale numerical values between 0 and 1.

- It ensures all features contribute equally
- Important for machine learning algorithms

MinMaxScaler is used for normalization.

```python
from sklearn.preprocessing import MinMaxScaler
```

### Applying Min-Max Scaling

The `fit_transform()` function:
- Fits the data (learns min and max values)
- Transforms data into a scaled range [0,1]

```python
scaler = MinMaxScaler()
df[['Age', 'EstimatedSalary']] = scaler.fit_transform(df[['Age', 'EstimatedSalary']])
```

```python
df.head()
```

## 11. Converting Categorical Data to Numerical

Categorical variable 'Gender' is converted into numerical format using One-Hot Encoding.

- `get_dummies()` creates binary columns (0 and 1)
- `drop_first=True` avoids redundancy (dummy variable trap)

```python
df = pd.get_dummies(df, columns=['Gender'], drop_first=True)
df.head()
```

## 12. Final Dataset

The final dataset is displayed along with updated data types after preprocessing.
